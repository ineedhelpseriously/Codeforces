```python
b=[1,3,5,7,9]
i=8
high=len(b)-1
low=0
while high>=low and i>=b[0]:
    mid=(high+low)//2
    if i>b[mid]:
        low=mid+1
    if i<b[mid]:
        high=mid-1
    if i==b[mid]:
        print(mid+1)
        break
    if i not in b and b[mid-1]<i and b[mid+1]>i:
        print(mid)
        break
else:
    print(0)
```

```python
import math
def isprime(n):
    if n<2: return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
n=10
m=math.sqrt(n)
isprime(m)
```
