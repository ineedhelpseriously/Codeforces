```python
import math
def isprime(n):
    if n<2: return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
```

```python
n=10
m=math.sqrt(n)
isprime(m)
```

```python
N = int(input())

is_prime = [True] * (N + 1)
is_prime[0] = is_prime[1] = False

for i in range(2, int(N**0.5) + 1):
    if is_prime[i]:
        for j in range(i * i, N + 1, i):
            is_prime[j] = False

for i in range(2, N + 1):
    if is_prime[i]:
        print(i, end=" ")
```

```python
n = int(input().strip())
mid = n // 2

for i in range(n):
    for j in range(n):
        if i == mid and j == mid:
            print('X', end='')
        elif i == j:
            print('\\', end='')
        elif i + j == n - 1:
            print('/', end='')
        else:
            print('*', end='')
    print()
```
